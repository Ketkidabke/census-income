# LINEAR REG

```python
# Target column : Hours per week
# Independent column : Education_num
```

```python
# Model building part

# seperate independent and dependent columns
# train test split
# build our model --- checking for classification or regression models ----> fit_transform
# make predictions
# evaluating our model
# if a classification model :  accuracy score, confusion matrix, classification report
# if a regression model : r2 score, adjusted r2 score, mse, rmse, mae
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
```

```python
lr=LinearRegression()   #making an object of linear regression class
```

```python
#As per the question independent variable is “education.num”.
x=data[['education_num']]
#dependent variable is “hours.per.week”
y=data['hours_per_week']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=1)
lr.fit(x_train,y_train)
y_pred=lr.predict(x_test)
```

```python
y_pred   #These are the predicted values
```

```python
y_test #these are actual values
```

```python
print('mean_squared_error :',mean_squared_error(y_test,y_pred))
```

```python
print('root_mean_squared_error :',mean_squared_error(y_test,y_pred, squared = False ))
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
import seaborn as sns
sns.regplot(x=y_test,y=y_pred,color='red')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
x=data[['occupation']]
# giving particular labels to the elements in my columns
L=LabelEncoder()
x=L.fit_transform(x)
x
```

```python
x=pd.DataFrame(x)
x
```

```python
y=data['annual_income'].replace('<=50K',0).replace('>50K',1)
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.35,random_state=1)
lo=LogisticRegression()
lo.fit(x_train,y_train)
y_pred=lo.predict(x_test)

print('confusion_matrix :')
print(confusion_matrix(y_pred,y_test))
print('accuracy_score :',accuracy_score(y_test,y_pred))
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.30,random_state=1)
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
y_pred=dt.predict(x_test)

print('confusion_matrix :')
print(confusion_matrix(y_pred,y_test))
print('accuracy_score :',accuracy_score(y_test,y_pred))
```
